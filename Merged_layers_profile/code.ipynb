```python
#!pip install textblob
from textblob import TextBlob
import pandas as pd
import csv

profile_data = pd.read_csv("final_dataset/final_dataset.csv", encoding="latin-1", quoting=csv.QUOTE_NONNUMERIC)
reviews = profile_data['review'].tolist()
sentiment_score = []
sentiment_subjectivity=[]
review_head_sentiment=[]
for rev in reviews:
    testimonial = TextBlob(str(rev))
    sentiment_score.append(testimonial.sentiment.polarity)
    sentiment_subjectivity.append(testimonial.sentiment.subjectivity)
```

```python
testimonial = TextBlob("I'm really sad with the product")
testimonial.sentiment.polarity
```

```python
testimonial = TextBlob("It was awesome! I've never had something like this before. Definitely a must buy!")
testimonial.sentiment.polarity
```

```python
profile_data['review_head_sentiment'] = sentiment_score
profile_data['review_sentiment'] = sentiment_score
profile_data['sentiment_subjectivity'] = sentiment_subjectivity
```
